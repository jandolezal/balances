# Bioenergy 2002-2018

Get key variables as csv export for 2018 (data from Eurostat's energy balances) and all years for Czechia (bioenergy, biofuels, solid biomass)

```python
import os
import datetime
import pandas as pd
import numpy as np
```

```python
csv_output_dir = datetime.datetime.today().strftime('%Y-%m-%d')

if not os.path.exists(csv_output_dir):
    os.mkdir(csv_output_dir)
```

```python
csvfile = os.path.join(os.path.abspath('data'), 'eurostat_2002_2018_tj.csv')

df = pd.read_csv(csvfile, decimal=',', index_col=['country', 'year', 'fuel'])
df
```

```python
#Â Key variables for bioenergy in 2018

bioenergy_2018 = df.loc[pd.IndexSlice[:,2018,'bioenergy'], :]
bioenergy_2018.to_csv(os.path.join(csv_output_dir, 'bioenergy_2018_tj.csv'), decimal=',')
bioenergy_2018
```

```python
#Â Key variables for biofuels in 2018

biofuels_2018 = df.loc[pd.IndexSlice[:,2018,'biofuels'], :]
biofuels_2018.to_csv(os.path.join(csv_output_dir, 'biofuels_2018_tj.csv'), decimal=',')
biofuels_2018
```

```python
#Â Key variables for solid biomass in 2018

solid_2018 = df.loc[pd.IndexSlice[:,2018,'solid_biomass'], :]
solid_2018.to_csv(os.path.join(csv_output_dir, 'solid_2018_tj.csv'), decimal=',')
solid_2018
```

```python
#Â Key variables for renewable municipal waste in 2018

mun_waste_2018 = df.loc[pd.IndexSlice[:,2018,'ren_mun_waste'], :]
mun_waste_2018.to_csv(os.path.join(csv_output_dir, 'ren_mun_waste_2018_tj.csv'), decimal=',')
mun_waste_2018
```

```python
consumption_bioenergy = df.loc[pd.IndexSlice[:,:,'bioenergy'], 'gross_inland_consumption']
```

```python
consumption_bioenergy.unstack('country')
```

```python
consumption_bioenergy.unstack('country').reset_index('fuel').plot(
    title='Gross inland consumption of bioenergy',
    ylabel='TJ',
    )
```

```python
dependency_bioenergy = df.loc[pd.IndexSlice[:,:,'bioenergy'], 'dependency']

dependency_bioenergy.unstack('country').reset_index('fuel').plot(
    title='Energy dependency of bioenergy',
    ylabel='Share of imports on gross inland consumption',
    )
```

```python
cz_bioenergy = df.loc[pd.IndexSlice['CZ',:,'bioenergy'], :]
cz_bioenergy.to_csv(os.path.join(csv_output_dir, 'cz_bioenergy_tj.csv'), decimal=',')
cz_bioenergy
```
